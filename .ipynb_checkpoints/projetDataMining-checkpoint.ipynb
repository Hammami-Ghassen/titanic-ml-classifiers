```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
####1
```

```python
data = pd.read_csv('train.csv')
data
```

```python
print("Colonnes disponibles :", data.columns.tolist())
```

```python
# Nettoyage 
colonnes_a_supprimer = ['PassengerId', 'Name', 'Ticket', 'Cabin']
colonnes_existantes = [col for col in colonnes_a_supprimer if col in data.columns]

if colonnes_existantes:
    data = data.drop(colonnes_existantes, axis=1)
else:
    print("Aucune des colonnes à supprimer n'a été trouvée")
```

```python
# Remplissage des valeurs manquantes
data['Age'] = data['Age'].fillna(data['Age'].median())
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
# Encodage des variables catégory
data = pd.get_dummies(data, columns=['Sex', 'Embarked'])
```

```python
### 2
```

```python
#Diviser les données
X = data.drop('Survived', axis=1)
y = data['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
### 3
```

```python
# Former les classificateurs
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
svm = SVC()
svm.fit(X_train, y_train)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
### 4
```

```python
# Prédictions
dt_pred = dt.predict(X_test)
svm_pred = svm.predict(X_test)
knn_pred = knn.predict(X_test)
```

```python
#### 5 
```

```python
#Évaluation
def evaluate(model, pred, name):
    accuracy = accuracy_score(y_test, pred)
    print(f"Précision de {name}: {accuracy:.2f}")
    
    cm = confusion_matrix(y_test, pred)
    plt.figure()
    sns.heatmap(cm, annot=True, fmt='d')
    plt.title(f"Matrice de confusion - {name}")
    plt.show()

evaluate(dt, dt_pred, "Arbre de Décision")
evaluate(svm, svm_pred, "SVM")
evaluate(knn, knn_pred, "KNN")
```

```python
### 6
```

```python
#Comparaison
results = pd.DataFrame({
    'Modèle': ['Arbre de Décision', 'SVM', 'KNN'],
    'Précision': [
        accuracy_score(y_test, dt_pred),
        accuracy_score(y_test, svm_pred),
        accuracy_score(y_test, knn_pred)
    ]
})


print("\nComparaison des modèles:")
print(results.sort_values(by='Précision', ascending=False))
```
